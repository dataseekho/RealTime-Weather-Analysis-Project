#### File reading in Java
```java
import java.io.BufferedReader;
import java.io.FileReader;
import java.io.IOException;

public class ReadFile {
    public static void main(String[] args) throws IOException{
        String fileContents = readEntireFile("./foo.txt");
    }

    private static String readEntireFile(String filename) throws IOException {
        FileReader in = new FileReader(filename);
        StringBuilder contents = new StringBuilder();
        char[] buffer = new char[4096];
        int read = 0;
        do {
            contents.append(buffer, 0, read);
            read = in.read(buffer);
        } while (read >= 0);
        return contents.toString();
    }
}
```

# **Scenario: Managing a contact list.**

* **List**: You might use a list to store the names of all your contacts. For
example, a list could include names like ["Alice", "Bob", "Charlie"].

* **Tuple**: You might use a tuple to store the contact information for a single person that shouldnt change, such as their phone number and email address: ("123-456-7890", "alice@example.com").

* **Dictionary**: You could use a dictionary to map each contacts name to their contact information. For example, {"Alice": ("123-456-7890", "alice@example.com"), "Bob": ("987-654-3210", "bob@example.com")}.

* **Set**: You might use a set to keep a collection of unique contact names, ensuring no duplicate names are stored. For example, a set might look like {"Alice", "Bob", "Charlie"}.



# **Sets: An unordered collection of unique objects.**
- Unique Elements: Sets automatically ensure that all elements are unique; duplicates are not allowed.

````python
# Create a list with duplicate names
contact_list = ["Alice", "Bob", "Alice", "Charlie", "Bob"]

# Convert the list to a set to remove duplicates
unique_contacts = set(contact_list)

print(unique_contacts)  # Output: {'Alice', 'Bob', 'Charlie'}
````

#### Dictionaries (dict)
* Unordered collection of key-to-value pairs
* dict elements can be accessed by key, but *not* by position
* Syntax for creating a dictionary is curly brackets --> my_dictionary = {}
    * you could declare it like this: my_dictionary = dict() --> but again, the curly brackets method is more common

````python
dict = {key:values}
````

````python
# A dictionary is an unordered mapping from keys to values
fruit_prices = {
    'apple': 0.65,
    'mango': 1.50,
    'strawberry': '$3/lb',
    'durian': 'unavailable',
    'student': ['student1':1]
}
````

````python
# Access the value associated with the 'student' key
student_data = fruit_prices['student']

# Access the value associated with the 'student1' key within the nested dictionary
student1_value = student_data['student1']

print(student1_value)  # Output: 1
````

````python
# Access the price of a mango
print(f"The price of a mango is: ${fruit_prices['mango']}")
````

````python
# Add a new entry for pears
fruit_prices['pear'] = 0.75
print(f"Updated dictionary: {fruit_prices}")
````

````python
# Example 1: Updating the price of an existing fruit
fruit_prices['apple'] = 0.70
print(f"Updated price of apple: ${fruit_prices['apple']}")
````

````python
# Example 2: Removing an entry from the dictionary
del fruit_prices['durian']
print(f"Dictionary after removing durian: {fruit_prices}")
````

## Classes
* A template for a particular kind of object
* A class defines the variables an object contains and what it can do with them
* To illustrate, let's define a `Student` class...
* Note: object-oriented programming can be a bit hard to understand at first, and we're moving quickly

````python
# Defining a simple Student class
class Student:
    # The __init__ method initializes the student with name and grade
    def __init__(self, name, grade):
        self.name = name  # Student's name
        self.grade = grade  # Student's grade

    # Method to show student details
    def show_info(self):
        return f"Name: {self.name}, Grade: {self.grade}"

    # Method to check if the student passed
    def has_passed(self):
        if self.grade >= 60:
            return f"{self.name} has passed."
        else:
            return f"{self.name} has not passed."

````

````python
name = input("Enter your name: ")
grade = int(input("Enter your grade: "))
````

````python
# Creating student objects
student1 = Student(name, grade)
#student1 = Student("Bob", 55)
````

````python
# Showing student info and checking if they passed
print(student1.show_info())
print(student1.has_passed())
````
